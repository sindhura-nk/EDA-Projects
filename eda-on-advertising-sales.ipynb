## Data Visualization

Dataset has only continuous features i.e quantitative data

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

Univariate Analysis

```python
con = ['TV','Radio','Newspaper']
for i in con:
    sns.histplot(df,x=i,kde=True)
    plt.title(f'Histogram for {i}')
    plt.show()
```

Bivariate Analysis

```python
for i in con:
    sns.scatterplot(df,x=i,y='Sales')
    plt.title(f'Scatterplot for {i}')
    plt.show()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
sns.pairplot(df)
```

## Observation:

    TV ads and Sales are highly correlated. Increase in TV ads can increase the number of sales
    Radio ads provide atleast 20 sales irrespective of number of radio ads.
    Sales are high when radio ads are above 40
    Newspaper ads ranging from 15-75 ads provide a minumum of 10 sales and maximum of 25sales
